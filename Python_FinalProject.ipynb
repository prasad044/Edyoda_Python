## You have to design a Food Ordering app for a restaurant

```python
class FoodItem:
    def __init__(self, food_id, name, quantity, price, discount, stock):
        self.food_id = food_id
        self.name = name
        self.quantity = quantity
        self.price = price
        self.discount = discount
        self.stock = stock

class RestaurantMenu:
    def __init__(self):
        self.menu = []

    def add_food_item(self, name, quantity, price, discount, stock):
        food_id = len(self.menu) + 1  # Generate FoodID automatically
        food_item = FoodItem(food_id, name, quantity, price, discount, stock)
        self.menu.append(food_item)
        print(f"Food item '{food_item.name}' added with FoodID {food_id}.")

    def edit_food_item(self, food_id, name, quantity, price, discount, stock):
        for item in self.menu:
            if item.food_id == food_id:
                item.name = name
                item.quantity = quantity
                item.price = price
                item.discount = discount
                item.stock = stock
                print(f"Food item {item.name} (FoodID: {food_id}) updated successfully.")
                return
        print(f"Food item with FoodID {food_id} not found.")

    def view_all_food_items(self):
        print("Restaurant Menu:")
        for item in self.menu:
            print(f"FoodID: {item.food_id}, Name: {item.name}, Quantity: {item.quantity}, Price: {item.price}, Discount: {item.discount}, Stock: {item.stock}")

    def remove_food_item(self, food_id):
        for item in self.menu:
            if item.food_id == food_id:
                self.menu.remove(item)
                print(f"Food item {item.name} (FoodID: {food_id}) removed successfully.")
                return
        print(f"Food item with FoodID {food_id} not found.")

# Initialize the restaurant menu
menu = RestaurantMenu()

while True:
    print("\nRestaurant Menu Management:")
    print("1. Add Food Item")
    print("2. Edit Food Item")
    print("3. View All Food Items")
    print("4. Remove Food Item")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        name = input("Enter the name of the food item: ")
        quantity = input("Enter the quantity (e.g., 100ml, 250gm, 4 pieces): ")
        price = float(input("Enter the price: "))
        discount = float(input("Enter the discount: "))
        stock = int(input("Enter the stock amount: "))
        menu.add_food_item(name, quantity, price, discount, stock)

    elif choice == "2":
        food_id = int(input("Enter the FoodID of the food item you want to edit: "))
        name = input("Enter the new name of the food item: ")
        quantity = input("Enter the new quantity (e.g., 100ml, 250gm, 4 pieces): ")
        price = float(input("Enter the new price: "))
        discount = float(input("Enter the new discount: "))
        stock = int(input("Enter the new stock amount: "))
        menu.edit_food_item(food_id, name, quantity, price, discount, stock)

    elif choice == "3":
        menu.view_all_food_items()

    elif choice == "4":
        food_id = int(input("Enter the FoodID of the food item you want to remove: "))
        menu.remove_food_item(food_id)

    elif choice == "5":
        print("Exiting the menu management system.")
        break

    else:
        print("Invalid choice. Please select a valid option.")
```

```python
class User:
    def __init__(self, full_name, phone_number, email, address, password):
        self.full_name = full_name
        self.phone_number = phone_number
        self.email = email
        self.address = address
        self.password = password
        self.orders = []

class FoodItem:
    def __init__(self, name, description, price):
        self.name = name
        self.description = description
        self.price = price

class Order:
    def __init__(self, user, items):
        self.user = user
        self.items = items

users = []
food_menu = [
    FoodItem("Tandoori Chicken", "4 pieces", 240),
    FoodItem("Vegan Burger", "1 Piece", 320),
    FoodItem("Truffle Cake", "500gm", 900)
]
orders = []

def register():
    full_name = input("Enter your Full Name: ")
    phone_number = input("Enter your Phone Number: ")
    email = input("Enter your Email: ")
    address = input("Enter your Address: ")
    password = input("Create a Password: ")

    user = User(full_name, phone_number, email, address, password)
    users.append(user)
    print("Registration Successful!")

def login():
    email = input("Enter your Email: ")
    password = input("Enter your Password: ")

    for user in users:
        if user.email == email and user.password == password:
            return user

    print("Login failed. Please check your credentials.")
    return None

def place_order(user):
    print("Food Menu:")
    for i, item in enumerate(food_menu, start=1):
        print(f"{i}. {item.name} ({item.description}) [INR {item.price}]")

    selected_items = []
    while True:
        try:
            item_numbers = input("Enter the numbers of the items you want to order (separated by commas): ")
            item_numbers = [int(x) for x in item_numbers.split(',')]
            selected_items = [food_menu[i - 1] for i in item_numbers]
            break
        except (ValueError, IndexError):
            print("Invalid input. Please enter valid item numbers.")

    total_price = sum(item.price for item in selected_items)
    print("\nSelected Items:")
    for item in selected_items:
        print(f"{item.name} ({item.description}) [INR {item.price}]")

    confirm_order = input(f"Total Price: INR {total_price}. Confirm order (yes/no): ").lower()
    if confirm_order == "yes":
        order = Order(user, selected_items)
        orders.append(order)
        print("Order placed successfully!")

def order_history(user):
    print("Order History:")
    for order in user.orders:
        print(f"Order Date: {order.date}")
        for item in order.items:
            print(f"{item.name} ({item.description}) [INR {item.price}]")
        print(f"Total Price: INR {sum(item.price for item in order.items)}")
        print("---------------")

def update_profile(user):
    print("Update Profile:")
    user.full_name = input(f"Full Name ({user.full_name}): ")
    user.phone_number = input(f"Phone Number ({user.phone_number}): ")
    user.email = input(f"Email ({user.email}): ")
    user.address = input(f"Address ({user.address}): ")
    user.password = input("Change Password: ")

def main():
    while True:
        print("\nOptions:")
        print("1. Register")
        print("2. Login")
        print("3. Place New Order")
        print("4. Order History")
        print("5. Update Profile")
        print("6. Quit")

        choice = input("Enter your choice: ")

        if choice == "1":
            register()
        elif choice == "2":
            logged_in_user = login()
            if logged_in_user:
                print(f"Welcome, {logged_in_user.full_name}!")
        elif choice == "3":
            if logged_in_user:
                place_order(logged_in_user)
            else:
                print("Please login first.")
        elif choice == "4":
            if logged_in_user:
                order_history(logged_in_user)
            else:
                print("Please login first.")
        elif choice == "5":
            if logged_in_user:
                update_profile(logged_in_user)
                print("Profile updated successfully!")
            else:
                print("Please login first.")
        elif choice == "6":
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```
